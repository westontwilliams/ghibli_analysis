# Interpretation (4)

Describe something interesting about your corpus that you discovered during the process of completing this assignment.

At a minumum, use 250 words, but you may use more. You may also add images if you'd like.

One of the more unexpected findings came from the hierarchical clustering of films using their combined sentiment and LDA topic profiles. Rather than cleanly separating by director, which I thought would at least be somewhat noticeable going into this,the dendrogram revealed that thematic and emotional content overlaps significantly across the Miyazaki/Takahata filmographies, far more so than I expected. Certain Miyazaki films clustered closer to Takahata's work than to other Miyazaki films, like in the case of Spirited Away and The Tale of the Princess Kaguya, suggesting that there's enough variation within each director's body of works to avoid there being a clear distrinction between them.

This is particularly visible when comparing films by release period. Earlier Miyazaki works like Nausicaä and Castle in the Sky share a heavier load of fear and anger in their emotion profiles. This is consistent with their apocalyptic and war-adjacent settings, while later films like Ponyo and Kiki's Delivery Service skew strongly toward joy and anticipation. Takahata's films, meanwhile, are more emotionally variable: Grave of the Fireflies is a devastating movie about the horror of war, but Only Yesterday and Princess Kaguya cluster close to some of Miyazaki's quieter, more domestic work.

Overall, the emotional tone of a Ghibli film is more dependent on its subject matter than who directed it. Regardless of a movie being created by Miyazaki or Takahata, if it deals with war or survival, it's going to get a higher score in the fear and anger emotions, and films about everyday life and growing up will score higher in joy and anticipation. My initial hypothesis that it would be easy to identify the director a movie solely by its script was inaccurate.
